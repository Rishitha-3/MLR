```python
###############################IMPORTING FILE################################# 
import pandas as pd
df = pd.read_csv("ToyataCorolla.csv")
df.head()
```

```python
#EDA --> boxplot, histogrmam, scatter plot, 

#shown in mlr models file
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
#numerical data
df_nom = df[['Age_08_04', 'KM', 'HP', 'Automatic', 'cc', 'Doors', 'Cylinders', 'Gears', 'Weight']]
```

```python
df_nom['Age_08_04'] = SS.fit_transform(df_nom[['Age_08_04']])
df_nom['KM'] = SS.fit_transform(df_nom[['KM']])
df_nom['HP'] = SS.fit_transform(df_nom[['HP']])
df_nom['Automatic'] = SS.fit_transform(df_nom[['Automatic']])
df_nom['cc'] = SS.fit_transform(df_nom[['cc']])
df_nom['Doors'] = SS.fit_transform(df_nom[['Doors']])
df_nom['Cylinders'] = SS.fit_transform(df_nom[['Cylinders']])
df_nom['Gears'] = SS.fit_transform(df_nom[['Gears']])
df_nom['Weight'] = SS.fit_transform(df_nom[['Weight']])

df_nom.head()
```

```python
#label encoding
from sklearn.preprocessing import  LabelEncoder
LE = LabelEncoder()
X_cat=LE.fit_transform(df["Fuel_Type"])
X_cat
```

```python
df_cat = pd.DataFrame(X_cat, columns=['Fuel Type '])
df_cat
```

```python
X_SS=pd.concat([df_nom,df_cat],axis=1)
X_SS
```

```python
Y = df["Price"]
Y.head()
```

```python
#Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_SS,Y,test_size=0.25)
```

```python
############################MODEL FITTING#######################################
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test  = model.predict(X_test)
```

```python
############################# data validation#########################################
from sklearn.metrics import mean_squared_error
train_rmse = mean_squared_error(Y_train,Y_pred_train,squared=False)
print("Training Error:",train_rmse)
test_rmse = mean_squared_error(Y_test,Y_pred_test,squared=False)
print("Test Error:",test_rmse)
```

```python
training_err = []
test_err = []

for i in range(1,101):
    X_train,X_test,Y_train,Y_test = train_test_split(X_SS,Y, test_size=0.25,random_state=i)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test  = model.predict(X_test)
    training_err.append(mean_squared_error(Y_train,Y_pred_train,squared=False))
    test_err.append(mean_squared_error(Y_test,Y_pred_test,squared=False))

import numpy as np
print("Cross validation - Training error:" ,np.mean(training_err).round(2))
print("Cross validation - Test error:" ,np.mean(test_err).round(2))
```

```python
LR_coef = model.coef_
LR_coef
```

```python
################################## LASSO REGRESSION #########################################
from sklearn.linear_model import Lasso
X_train,X_test,Y_train,Y_test = train_test_split(X_SS,Y,test_size=0.25,random_state=100)

model = Lasso(alpha=1) 

model.fit(X_train,Y_train)

model.intercept_
coeff = pd.concat([pd.DataFrame(list(X_SS)),pd.DataFrame(LR_coef),pd.DataFrame(model.coef_)],axis=1)
coeff
```
